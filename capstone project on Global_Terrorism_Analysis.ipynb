##**Countrywise analysis**

A function on each countrywise analysis, In that Terror attacks on top 10 cities of country,Active terrorist groups in respective country and Total casualities from 1970 to 2017.

```python
def country_analysis(data , Country=None, line_color=None , palette=None , title =None):
    plt.figure(figsize = (17,10))
    #Plot 1
    plt.subplot(2,2,1)
    sns.countplot(y = data.loc[data['Country'] == Country]['City'] ,\
                  order = data.loc[data.Country == Country].City.value_counts()[:10].index,\
                  data = data ,\
                  palette = sns.color_palette(palette,10),\
                  edgecolor = sns.color_palette('dark',5))
    plt.xticks(rotation = 90)
    plt.title(f"Terror attacks on Top 10 cities of {Country}",weight = 'bold' ,fontsize = 15 ,loc = 'center')
    plt.xlabel('Count',fontsize = 15)
    plt.ylabel('City',fontsize = 15)
    #Plot 2
    plt.subplot(2,2,3)
    sns.countplot(x = data.query('Country == @Country').Gang_Name ,\
                  data = data ,\
                  order = data.query('Country == @Country').Gang_Name.value_counts()[:10].index[1:],\
                  palette = sns.color_palette(palette,7),\
                  edgecolor = sns.color_palette('dark',5))
    plt.title(f"Terrorist Groups active/were active in {Country}",weight = 'bold' ,fontsize = 15 , loc = 'center')
    plt.xticks(rotation =90)
    plt.xlabel('Group',fontsize = 15)
    plt.ylabel('Attack Count',fontsize = 15)
    

    # Plot 3
    plt.subplot(2,2,2)
    data.query('Country == @Country').groupby(['Year'])['Casualties'].sum().plot(xticks = data.loc[data.Country == Country].groupby(['Year'])['Casualties'].sum().index, rot=90 , color = line_color)
    plt.title(f"Total casualities from (1970 - 2017) in {Country}", fontdict = {'fontsize' : 15 , 'weight': 'bold' } , loc = 'center')
    plt.ylabel("Casualities" ,fontsize = 15)
    plt.xlabel('Year',fontsize = 15)

    #plot 4
    plt.subplot(2,2,4)
    sns.countplot(y = data.loc[data['Country'] == Country]['Main_Target'] ,\
                  order = data.loc[data.Country == Country].Main_Target.value_counts()[:10].index,\
                  data = data ,\
                  palette = sns.color_palette(palette,10),\
                  edgecolor = sns.color_palette('dark',5))
    plt.title(f"Most Attacked Targets by Terrorism Groups in {Country}",fontsize=15,weight='bold',loc='center')
    plt.xlabel('Target type',fontsize=15)
    plt.ylabel('Number of Attacks',fontsize=15)

    plt.subplots_adjust(wspace = 1 ,hspace =0.5 ,top = 0.5 , bottom = 0.1) 
    plt.tight_layout()  
    plt.show()
```

###Country Iraq

```python
# Analysis of country Iraq
country_analysis(new_df, Country ='Iraq', palette = 'inferno', line_color = 'crimson')
```
